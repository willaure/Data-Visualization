```python
# Visualizing the Distribution of Data¬∂
# Histograms
# Plot a histogram of CompTotal (Total Compensation).
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the style and context for better readability
sns.set_theme(style="whitegrid", context="notebook", font_scale=1.2)
plt.figure(figsize=(14, 8))

# 1. Data Cleaning
comp_data = df['CompTotal'].dropna()
comp_data = comp_data[comp_data > 0]  # Remove non-positive values

# 2. Plot the histogram with KDE
ax = sns.histplot(
    comp_data,
    bins=50,
    color='#3498db',  # Professional blue
    edgecolor='white',
    linewidth=0.5,
    kde=True,  # Kernel Density Estimate overlay
    alpha=0.8,  # Slightly transparent for better KDE visibility
    log_scale=True  # Use log scale for better distribution visibility
)

# 3. Add statistical markers
mean_val = comp_data.mean()
median_val = comp_data.median()
plt.axvline(mean_val, color='#e74c3c', linestyle='--', linewidth=2.5, label=f'Mean (${mean_val:,.0f})')
plt.axvline(median_val, color='#2ecc71', linestyle='-', linewidth=2.5, label=f'Median (${median_val:,.0f})')

# 4. Improve x-axis labels (log-scaled)
ax.set_xticks([1e4, 2.5e4, 5e4, 1e5, 2.5e5, 5e5, 1e6, 2.5e6, 5e6, 1e7])
ax.set_xticklabels(
    ['$10k', '$25k', '$50k', '$100k', '$250k', '$500k', '$1M', '$2.5M', '$5M', '$10M'],
    rotation=45
)

# 5. Add annotations and titles
plt.title('Distribution of Total Compensation (Log Scale)', fontsize=18, pad=20, weight='bold')
plt.xlabel('Total Compensation (USD)', fontsize=14, labelpad=10)
plt.ylabel('Number of Respondents', fontsize=14, labelpad=10)
plt.legend(frameon=True, framealpha=1, facecolor='white')

# 6. Add summary statistics in a box
stats_text = f"""
Key Statistics:
- Mean: ${mean_val:,.0f}
- Median: ${median_val:,.0f}
- Std Dev: ${comp_data.std():,.0f}
- Range: ${comp_data.min():,.0f} to ${comp_data.max():,.0f}
- Total Respondents: {len(comp_data):,}
"""
plt.annotate(
    stats_text,
    xy=(0.95, 0.5),
    xycoords='axes fraction',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.9),
    fontsize=12,
    ha='right'
)

# 7. Adjust grid and layout
plt.grid(axis='y', alpha=0.3)
plt.grid(axis='x', alpha=0.1)  # Light grid for log scale
plt.tight_layout()
plt.show()
```

```python
# Box Plots
# Plot a box plot of Age.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the style
sns.set_theme(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(10, 6))

# 1. Data Preparation - Convert age ranges to numeric if needed
if df['Age'].dtype == 'object':
    # Map categorical age ranges to midpoint values
    age_mapping = {
        'Under 18 years old': 17,
        '18-24 years old': 21,
        '25-34 years old': 29.5,
        '35-44 years old': 39.5,
        '45-54 years old': 49.5,
        '55-64 years old': 59.5,
        '65 years or older': 70
    }
    age_data = df['Age'].map(age_mapping).dropna()
else:
    age_data = df['Age'].dropna()

# 2. Create the box plot
ax = sns.boxplot(
    x=age_data,
    color='#3498db',
    width=0.4,
    linewidth=2.5,
    flierprops=dict(
        marker='o',
        markerfacecolor='#e74c3c',
        markersize=8,
        alpha=0.6
    )
)

# 3. Calculate and annotate statistics
stats = age_data.describe()
median = stats['50%'] if '50%' in stats.index else stats['median']
q1 = stats['25%'] if '25%' in stats.index else stats['first quartile']
q3 = stats['75%'] if '75%' in stats.index else stats['third quartile']

plt.text(0.05, 0.95,
         f"Median: {median:.1f}\nQ1: {q1:.1f}\nQ3: {q3:.1f}\nIQR: {q3-q1:.1f}",
         transform=ax.transAxes,
         bbox=dict(facecolor='white', alpha=0.9),
         fontsize=12,
         verticalalignment='top')

# 4. Customize the plot
title = 'Distribution of Age' + (' (Numeric Midpoints)' if df['Age'].dtype == 'object' else '')
plt.title(title, fontsize=16, pad=20, weight='bold')
plt.xlabel('Age' + (' (Midpoint of Range)' if df['Age'].dtype == 'object' else ''),
          fontsize=14, labelpad=10)
plt.grid(axis='x', alpha=0.3)
plt.yticks([])

plt.tight_layout()
plt.show()

# Print additional stats
print("Age Distribution Statistics:")
print("-" * 30)
print(f"Min age: {stats['min']:.1f}")
print(f"Max age: {stats['max']:.1f}")
print(f"Mean age: {stats['mean']:.1f}")
print(f"Standard deviation: {stats['std']:.1f}")
```

```python
# Bubble Plots
# Create a bubble plot of TimeSearching and Frustration using the Age column as the bubble size.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('survey-data.csv')

# Convert Age to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Drop rows with missing values
df = df.dropna(subset=['TimeSearching', 'Frustration', 'Age'])

# Normalize Age for bubble size
df['Age_normalized'] = df['Age'] / df['Age'].max() * 800  # Increased size scale

# Create the bubble plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='TimeSearching',
    y='Frustration',
    size='Age_normalized',
    sizes=(50, 800),  # Bigger bubbles
    hue='Age',        # Optional: color gradient by Age
    palette='viridis',
    alpha=0.7,
    edgecolor='black',  # Border for contrast
    linewidth=0.5,
    legend=False
)

plt.title('Bubble Plot: TimeSearching vs Frustration (Bubble Size = Age)')
plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Visualizing Composition of Data
# Pie Charts
# Create a pie chart of the top 5 databases(DatabaseWantToWorkWith) that respondents wish to learn next year.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('survey-data.csv')

# Drop missing values in the target column
db_data = df['DatabaseWantToWorkWith'].dropna()

# Split entries by comma and explode into individual rows
all_dbs = db_data.str.split(',').explode().str.strip()

# Count occurrences and select top 5
top_dbs = all_dbs.value_counts().head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_dbs,
    labels=top_dbs.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel1.colors
)

plt.title('Top 5 Databases Respondents Want to Work With Next Year')
plt.axis('equal')  # Ensures pie is circular
plt.show()
```

```python
# Stacked Charts
# Create a stacked bar chart of median TimeSearching and TimeAnswering for the age group 30 to 35.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('survey-data.csv')

# Convert numeric columns to appropriate types
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['TimeSearching'] = pd.to_numeric(df['TimeSearching'], errors='coerce')
df['TimeAnswering'] = pd.to_numeric(df['TimeAnswering'], errors='coerce')

# Filter for age group 30 to 35
age_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median values
median_searching = age_filtered['TimeSearching'].median()
median_answering = age_filtered['TimeAnswering'].median()

# Plot stacked bar chart
plt.figure(figsize=(6, 6))
plt.bar('Age 30‚Äì35', median_searching, label='Time Searching')
plt.bar('Age 30‚Äì35', median_answering, bottom=median_searching, label='Time Answering')

plt.ylabel('Median Time (minutes)')
plt.title('Median Time Searching and Answering (Age 30‚Äì35)')
plt.legend()
plt.show()
```

```python
# Visualizing Comparison of Data¬∂
# Line Chart
# Plot the median CompTotal for all ages from 45 to 60.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('survey-data.csv')

# Convert relevant columns to numeric (handle non-numeric values gracefully)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

# Filter for ages 45 to 60
age_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Group by Age and compute median CompTotal
median_comp_by_age = age_filtered.groupby('Age')['CompTotal'].median()

# Plot the line chart
plt.figure(figsize=(10, 5))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='teal')
plt.title('Median Compensation (CompTotal) by Age (45‚Äì60)')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart
# Create a horizontal bar chart using the MainBranch column.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('survey-data.csv')

# Count the occurrences of each category in the MainBranch column
mainbranch_counts = df['MainBranch'].value_counts()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Distribution of MainBranch Responses')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.gca().invert_yaxis()  # To display the largest value at the top
plt.tight_layout()
plt.show()
```
